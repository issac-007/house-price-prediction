```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Read CSV file
data = pd.read_csv("Downloads\house_price.csv")

# Convert to NumPy arrays
X = data[["SquareFeet", "Cent", "Bedrooms"]].values
y = data["PriceLakhs"].values

# Train model
model = LinearRegression()
model.fit(X, y)

print("===== House Price Prediction =====")

# Get user input
square_feet = float(input("Enter Square Feet: "))
cent = float(input("Enter Land Area (Cent): "))
bedrooms = int(input("Enter Number of Bedrooms: "))

# Predict price
predicted_price = model.predict([[square_feet, cent, bedrooms]])

print("\nPredicted House Price =", round(predicted_price[0], 2), "Lakhs")
```
